```python
import pandas as pd
txt_files = ["../data/datatraining.txt", "../data/datatest.txt", "../data/datatest2.txt"]
dfs = [pd.read_csv(f) for f in txt_files]

df_all = pd.concat(dfs, ignore_index=True)

if "date" in df_all.columns:
    df_all["date"] = pd.to_datetime(df_all["date"])
    df_all = df_all.sort_values("date").reset_index(drop=True)

df_all.to_csv("../data/occupancy_full.csv", index=False)

print("Saved as occupancy_full.csv")
```
